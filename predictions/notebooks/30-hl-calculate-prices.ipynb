```python
asteroid_null_predictions.index = asteroid_null.index
```

```python
asteroid_null_predictions.head(10)
```

```python
def nan_to_empty(row):
    return {x : '' if np.isreal(y) and np.isnan(y) else y for x,y in row.to_dict().items()}
```

```python
for spec_col in asteroid_null_predictions.columns:
    spec = spec_col.split('.')[1]
    print(spec)
    asteroid_null['spec'] = spec    
        
    asteroid_null['price'] = asteroid_null.apply(lambda row: scoring.price(nan_to_empty(row))[0], axis=1)
    
    asteroid_null['profit_'+spec] = asteroid_null.apply(lambda row: scoring.profit(nan_to_empty(row)), axis=1)
        
    print(asteroid_null['profit_'+spec].describe())
```

```python
profits_df = asteroid_null.merge(asteroid_null_predictions,left_index=True,right_index=True)
profits_df.shape
```

```python
def expected_profit(row):
    value = 0
    for spec_col in asteroid_null_predictions.columns:
        profit_col = 'profit_'+ spec_col.split('.')[1]
        value += row[profit_col]*row[spec_col]
        
    return value

profits_df['expected_profit'] = profits_df.apply(expected_profit, axis=1)
```

```python
profits_df.sort_values(by='expected_profit',ascending=False).head(10)
```

```python
profits_df[ ~np.isnan(profits_df['albedo'])& ~np.isnan(profits_df['diameter']) & ~np.isnan(profits_df['dv'])].sort_values(by='expected_profit',ascending=False).head(10)
```
