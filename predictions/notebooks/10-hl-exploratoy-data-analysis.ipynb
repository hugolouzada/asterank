```python
import pymongo
from pymongo import MongoClient
import pandas as pd
from pydqc import infer_schema, data_summary
```

```python
client = MongoClient('localhost', 27017)
db = client['asterank']
```

```python
asteroids = db['asteroids']
```

```python
asteroids.count()
```

```python
doc=asteroids.find_one();
keys =[key for key in doc]
print(keys)
```

```python
# query de uma 1 coluna
pd.DataFrame(asteroids.find({},{'_id':0, "epoch_cal" : 1}))
```

```python
asteroids_df = pd.DataFrame(asteroids.find({})).drop(['id','_id'],axis=1)
infer_schema.infer_schema(asteroids_df, 
                              "asteroids_20190414", 
                              output_root='../reports/data_quality', 
                              sample_size=1.0, 
                              type_threshold=0.5, 
                              n_jobs=4, 
                              base_schema=None)
```

```python
data_schema = pd.read_excel('../reports/data_quality/data_schema_asteroids_20190414.xlsx')
data_schema.head(100)
```

```python
data_summary.data_summary(data_schema, 
                          asteroids_df, 
                          "asteroids_20190414", 
                          sample_size=1.0,
                          output_root='../reports/data_quality',
                          n_jobs=4)
```

**Note**: It looks like the values of spec and spec_T that are filled in are just a small fraction of the asteroids. Maybe more complicated analysis are not possible, or maybe a data augmentation procedure is necessary.
